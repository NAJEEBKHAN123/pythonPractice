### Looping on groups

```python
movies
```

```python
df = pd.DataFrame(columns=movies.columns)

for group, data in genres: 
    top = data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]
    df = pd.concat([df, top], ignore_index=True)

print(df)
```

```python
#find numbers of movies starting with A for each group
def foo(group):
    return group['Series_Title'].str.startswith('A').sum()

res = genres.apply(foo)
print(res)

```

```python
#find ranking of each movies in the group according to IMDB Rating

def rank_movie(group):
    group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
    return group

genres.apply(rank_movie)
```
