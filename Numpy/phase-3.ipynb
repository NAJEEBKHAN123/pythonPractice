```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("Sales Data Array:")
print(sales_data)
```

```python
print("Row 1 :", sales_data[0])
```

```python
print("Row 1 without first col ", sales_data[0, 1:])
```

```python
print("Row 1 without first col ", sales_data[:, 1:])
```

```python
print('SHAPE of sale data: ', sales_data.shape)
```

```python
print("First 3 rows: ", sales_data[:3, 1:])
```

```python
print(sales_data[:, 1:2])
```

```python
print('total sales of paradise biryani: ', sales_data[0, 1:].sum())
```

```python
print('max sales of paradise biryani: ', sales_data[0, 1:].max())
```

```python
print('min sales of paradise biryani: ', sales_data[0, 1:].min())
```

```python
print('avg sales of paradise biryani: ', sales_data[0, 1:].mean())
```

```python
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumsum)

plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title('Average Cumulative Sales Over Quarters')
plt.xlabel("Years")
plt.ylabel("Average Sales")
plt.grid()
plt.show()

```
